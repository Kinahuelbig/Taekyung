```
import requests 
url = 'https://www.space.com/news'
response = requests.get(url)
if (response.ok):
  data = response.text
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
list = soup.find('div')
print(list)
```
